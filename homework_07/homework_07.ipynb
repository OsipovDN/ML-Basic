### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
import random
array_min_max = np.random.randint(random.randint(0, 100), random.randint(100, 1000), size = (4, 3, 2))

print(array_min_max)
print(array_min_max.min())
print(array_min_max.max())
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
first_array = np.random.randint(random.randint(0, 100), random.randint(100, 1000), size = (6, 4))
second_array = np.random.randint(random.randint(0, 100), random.randint(100, 1000), size = (4, 3))

result = first_array@second_array
print(result)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
array = np.random.randint(random.randint(0, 100), random.randint(100, 1000), size = (7, 7))
normalized = (array - array.min()) / (array.max() - array.min())
# print(array)
print(f"Среднее значение : {array.mean():.2f}")
print(f"Стандартное отклонение : {array.std():.2f}")
np.set_printoptions(precision = 2)
print(f"Нормализованная матрица: {normalized}")
```

### Посмотрите на распределение числовых признаков

```python
fig, axes = plt.subplots(1, 3, figsize = (15, 5))

sns.histplot(tips['total_bill'], kde = True, ax=axes[0], color = 'skyblue')
axes[0].set_title('Распределение Total Bill')

sns.histplot(tips['tip'], kde = True, ax = axes[1], color = 'salmon')
axes[1].set_title('Распределение Tip')

sns.histplot(tips['size'], discrete = True, ax = axes[2], color = 'green')
axes[2].set_title('Распределение Size (компания)')

plt.tight_layout()
plt.show()
```

### Найдите максимальное значение 'total_bill'

```python
print(f"Максимальное значение чаевых: {tips["total_bill"].max()}")
```

### Найдите количество курящих людей

```python
print(f"Колличество курящих: {(tips['smoker'] == 'Yes').sum()}")
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
print(f"Средний сумма чека в зависимости от дня недели: {tips.groupby('day')['total_bill'].mean()}")
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
median = tips['total_bill'].median()
expensive_bills = tips[tips['total_bill'] > median]
avg_tip_by_sex = expensive_bills.groupby('sex')['tip'].mean()

print(f"Медиана total_bill: {median:.2f}")
print("\nСредние чаевые при счете выше медианы:")
print(avg_tip_by_sex)
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker'] = tips['smoker'].map({'No': 0, 'Yes': 1})
print(f"{tips.head(50)}")
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips_for_plot = sns.load_dataset('tips')

sns.histplot(data = tips_for_plot, x ='total_bill', kde = True)
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
sns.scatterplot(data = tips_for_plot, x = 'total_bill', y = 'tip')

plt.title('Взаимосвязь между чеком и чаевыми')
plt.xlabel('Общая сумма чека')
plt.ylabel('Размер чаевых')
plt.show()
```

### Постройте pairplot

```python
sns.pairplot(tips_for_plot)

plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
sns.boxplot(data = tips_for_plot, x = 'day', y = 'total_bill')

plt.title('Распределение суммы счета в зависиости от дням недели')
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
sns.displot(data = tips_for_plot, x = 'tip', col = 'time', kde = True)

plt.plot()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
sns.relplot(data = tips_for_plot, x = 'total_bill', y = 'tip', col = 'sex', hue = 'smoker', kind = 'scatter')

plt.show()
```
