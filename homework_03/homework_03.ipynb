## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
def snake_case(old_name: str) -> str:
    new_name = ''
    new_name += old_name[0].lower()
    for ch in old_name[1:]:
        if ch.isupper():
            new_name += '_'
        new_name += ch.lower()
    return new_name


def camel_case(old_name: str) -> str:
    new_name = ''
    end_word = False
    new_name += old_name[0].upper()
    for ch in old_name[1:]:
        if ch == '_':
            end_word = True
            continue
        if end_word:
            end_word = False
            new_name += ch.upper()
        else:
            new_name += ch
    return new_name


def change_stile(text: str) -> str:
    if text[0].isupper():
        #Примитивно, но предположим, что если первый символ в верхнем регистре, то нам передали PascalCase
        return snake_case(text)
    else:
        return camel_case(text)


name = input("Введите наименование метода или функции: ")
new_str = change_stile(name)
print (f'"{name}" -> "{new_str}"')
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
def checkdata(day: int, month: int, year: int) -> bool:
    if not 0 < day <= 31:
        return False
    if not 0 < month <= 12:
        return False
    if not month in [1, 3, 5, 7, 8, 10, 12]:
        return False
    elif month in [4, 6, 9, 11]:
        if day == 31:
            return False
    else:
        if year % 4 == 0:
            if day > 29:
                return False
        else:
            if day > 28:
                return False
    return True


date = input("Введите дату в формате день.месяц.год:")
day, month, year = date.split(sep = '.')
try:
    day = int(day)
    month = int(month)
    year = int(year)
except ValueError:
    print("Введено не числовое значение даты")
print(f'{date} -> {"True" if checkdata(day, month, year) else "False"}')
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def is_prime(num: int) -> bool:
    if num > 1:
         for i in range(2, num):
            if num % i == 0:
                return False
    else:
       return False
    return True


try:
    number = int(input("Введите число для проверки на простоту: "))
    print(f'"{number}" -> {"простое число" if is_prime(number) else 'не прстое число'}')
except ValueError:
    print("Введенное значение не является числом")
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
data_map = {}
while True:
    user_data = input('Введите имя, фамилию, возраст и id пользователя через пробел')
    if user_data == '':
        break
    user = list(user_data.split())
    if len(user) != 4:
        print('некоректный ввод. Недостаточно данных. Попробуйте снова')
        continue
    if not (18 <= int(user[2]) <= 60):
        print('Введен некоектный возраст. Попробуйте снова')
        continue
    id = user[-1].zfill(8)
    if id in data_map:
        print(f'Позьзователь с "{id}" уже существует. Попробуйте снова')
        continue
    user[:2] = [item.title() for item in user[:2] if not item.istitle()]
    data_map[id] = tuple(user[:3])
print(data_map)
```
