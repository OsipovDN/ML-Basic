## Задание 1:
    Пользователь в бесконечном цикле вводит числа через консоль до тех пор, пока не введёт что-то, отличное от числа (после этого цикл завершается). В конце программа должна вывести сумму всех введённых чисел.

```python
sum = 0.0
while True:
    try:
        sum += float(input("Введите число: ",))
    except ValueError:
        print("Ввод завершен")
        break
print(f"Общая сумма введенных числе: {sum}")
```

## Задание 2:
    Пользователь вводит строку через консоль (состоящую только из букв в разном регистре). Программа должна создать новую строку, содержащую только те буквы из исходной строки, которые были в верхнем регистре.

```python
result = ''
input_str = input("Введите строку в разном регистре: ",)
for ch in input_str:
    if ch.isupper():
        result += ch
print(f'Введенная строка "{input_str}" результирующая строка: "{result}" ')
```

## Задание 3:
Классическая игра FizzBuzz. Программа в цикле выводит в консоль числа от 1 до 100.

<br>

**Примеры:**
* Если число кратно 3, вместо него выводится Fizz.
* Если кратно 5 — Buzz.
* Если кратно и 3, и 5 — FizzBuzz.

```python
for i in range(1, 100):
    if i % 3 == 0 and i % 5 == 0:
        print (f"{i} -> FizzBuzz")
    elif i % 3 == 0:
        print (f"{i} -> Fizz")
    elif i % 5 == 0:
         print (f"{i} -> Buzz")
```

## Задание 3: Плитка шоколада

<br>

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
number = [
    1000, 900, 500, 400,
    100, 90, 50, 40,
    10, 9, 5, 4,
    1
    ]
syb = [
    "M", "CM", "D", "CD",
    "C", "XC", "L", "XL",
    "X", "IX", "V", "IV",
    "I"
    ]
in_count = input("Введите число: ",)
check_str = in_count.replace(' ', '')
if check_str.isdigit():
    value = int(in_count)
    roman_num = ''
    i = 0
    while  value > 0:
        for _ in range(value // number[i]):
            roman_num += syb[i]
            value -= number[i]
        i += 1
    print(f"{in_count} -> {roman_num}")
else:
    print(f"Введенное значение {in_count} не соответствует требованиям")
```

## Задание 5: Проверка на вещественное число

* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`

```python
in_count = input("Введите число: ",)
in_count.strip()
result = False
point_count = 0

for i in range(0, len(in_count)):
    ch = in_count[i]
    if "0" <= ch <= "9":
        result =True
    elif ch == '.' or ch == ',':
        point_count += 1
        print(point_count)
        if point_count > 1:
            result = False
            break
        result = True
    else:
        result = False
        break
print(f"{in_count} -> {result}")
```
