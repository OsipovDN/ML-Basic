## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
in_number = input("Введите пятизначное число: ",)
in_number.strip()
if len(in_number) == 5 and in_number.isdigit():
    result = in_number[0]+in_number[3:0:-1]+in_number[-1]
    print (f"{in_number} -> {result}")
else :
    print(f"Введенное значение {in_number} не соответствует требованиям")
```

## Задание 2: Подсчет выходных до отпуска

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
day = input("Введите колличество дней до отпуска: ",)
day.strip()
try:
    number = int(day)
    if day.isdigit():
        weekend_day_count = (number // 7)*2
        remainder = number % 7
        if remainder == 6:
            weekend_day_count += 1
        elif remainder >= 7:
            weekend_day_count += 2
        print(f"{day} -> {weekend_day_count}")
    else :
        print(f"Введенное значение {day} не соответствует требованиям")
except value_error :
    print(f"Значение {day} не является целочисленным")
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
ln, wh, count = input("Введите длину, ширину и размер куска через пробел: ",).split()
ln = int(ln)
wh = int(wh)
count = int(count)
print(f"{ln}, {wh}, {count} -> {(count % ln == 0 or count % wh == 0) and (count < ln * wh)}")
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
number = [
    1000, 900, 500, 400,
    100, 90, 50, 40,
    10, 9, 5, 4,
    1
    ]
syb = [
    "M", "CM", "D", "CD",
    "C", "XC", "L", "XL",
    "X", "IX", "V", "IV",
    "I"
    ]
in_count = input("Введите число: ",)
in_count.strip()
if in_count.isdigit():
    value = int(in_count)
    roman_num = ''
    i = 0
    while  value > 0:
        for _ in range(value // number[i]):
            roman_num += syb[i]
            value -= number[i]
        i += 1
    print(f"{in_count} -> {roman_num}")
else:
    print(f"Введенное значение {in_count} не соответствует требованиям")
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
in_count = input("Введите число: ",)
in_count.strip()
result = False
point_count = 0

for i in range(0, len(in_count)):
    ch = in_count[i]
    if "0" <= ch <= "9":
        result =True
    elif ch == '.' or ch == ',':
        point_count += 1
        print(point_count)
        if point_count > 1:
            result = False
            break
        result = True
    else:
        result = False
        break
print(f"{in_count} -> {result}")
```
