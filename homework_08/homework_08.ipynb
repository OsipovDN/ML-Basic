$$

x & 2 & 3 \\
-1 & y & 4

## Задание 6.



* Запишите матрично-векторное уравнение $Ax = b$ в виде системы линейных уравнений.

* Решите его любым способом, найдя **вектор** $x$. Решение опишите.

$$
A =
 \begin{pmatrix}
  1 & 2 & 4 \\
  2 & 1 & -1 \\
  1 & -1 & 1 \\
 \end{pmatrix}
 $$

 $$
 b =
  \begin{pmatrix}
  3 \\
  -9 \\
  0 \\
 \end{pmatrix}
 $$

**№6 Решение**<br>
Решим методом Гаусса
1. Запишем систему в виде расширенной матрицы $(A|b)$ и приведем к треугольному виду:

$
\begin{pmatrix}
1 & 2 & 4 & \big| & 3 \\
2 & 1 & -1 & \big| & -9 \\
1 & -1 & 1 & \big| & 0
\end{pmatrix}
$

2. Обнуляем элементы под главной диагональю.

* Из второй строки ($II$) вычитаем первую ($I$), умноженную на 2: $II \to II - 2 \cdot I$
* Из третьей строки ($III$) вычитаем первую строку ($I$): $III \to III - I$

$
\begin{pmatrix}
1 & 2 & 4 & \big| & 3 \\
0 & -3 & -9 & \big| & -15 \\
0 & -3 & -3 & \big| & -3
\end{pmatrix}
$

* Разделим вторую строку ($II$) на $-3$, чтобы упростить вычисления: $II \to II \cdot \left(-\frac{1}{3}\right)$

$
\begin{pmatrix}
1 & 2 & 4 & \big| & 3 \\
0 & 1 & 3 & \big| & 5 \\
0 & -3 & -3 & \big| & -3
\end{pmatrix}
$

* К третьей строке ($III$) прибавляем вторую ($II$), умноженную на 3: $III \to III + 3 \cdot II$

$
\begin{pmatrix}
1 & 2 & 4 & \big| & 3 \\
0 & 1 & 3 & \big| & 5 \\
0 & 0 & 6 & \big| & 12
\end{pmatrix}
$

* Запишем матрицу в виде системы линейных уравнений

$
\begin{cases}
x_1 + 2x_2 + 4x_3 = 3 \\
x_2 + 3x_3 = 5 \\
6x_3 = 12
\end{cases}
$

3. Теперь найдем значение $x_3$ из третьей строки и подставим во вторую для того что бы найти $x_2$ и в итоге $x_1$ двигаясь снизу вверх и подставляя значения:

* Из третьего уравнения:

$6x_3 = 12 \implies x_3 = 2$

* Подставляем $x_3 = 2$ во второе уравнение:

$x_2 + 3 \cdot 2 = 5 \implies x_2 + 6 = 5 \implies x_2 = -1$

* Подставляем $x_3 = 2$ и $x_2 = -1$ в первое уравнение:

$x_1 + 2 \cdot (-1) + 4 \cdot 2 = 3$
$x_1 - 2 + 8 = 3$
$x_1 + 6 = 3 \implies x_1 = -3$

**Ответ<br>**

Искомый вектор $x = \begin{pmatrix} x_1 \\ x_2 \\ x_3 \end{pmatrix}$ равен:

$
x = \begin{pmatrix} -3 \\ -1 \\ 2 \end{pmatrix}
$

**7-2:** Запишите функцию вероятности (CDF) этого распределения.








**№7.1 Решение**<br>
* Найдем значение y при значении случайной величины 0,5:<br>

* Найдем вероятность попадания в исходный интервал.<br>
    Вероятностью попадания случайной величины в интервал от 0 до 0,5 внутри         непрерывного распределения является площадь этого участка.

**№7.2 Решение**<br>
*  Найдем значение $F(x)$ при значении $x<0$<br>
$F(x) = 0$ т.к. значнеие находится вне диапазона<br>
*  Найдем значение $F(x)$ при значении $0<=x<=1$ т.к. максимальная вероятность (площадь) не должна быть больше одного<br>
    Для этого найдем первообразную исходной функции<br>
$F(x) = \int_{0}^{x} 2t \, dt = \left. t^2 \right|_{0}^{x} = x^2 - 0 = x^2$<br>
* Найдем значение $F(x)$ при значении $x>1$<br>
$F(x) = \int_{-\infty}^{x} F(t) \, dt = \int_{0}^{1} 2t \, dt + \int_{1}^{x} 0 \, dt = \left. t^2 \right|_{0}^{1} + 0 = (1^2 - 0^2) = 1$<br>
