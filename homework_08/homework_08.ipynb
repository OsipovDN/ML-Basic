### Задание 3.

Дано выражение:

$$
3 \cdot \begin{pmatrix}
x & 2 & 3 \\
-1 & y & 4
\end{pmatrix}
+ 2 \cdot \begin{pmatrix}
1 & 2 & -5 \\
2 & -6 & z
\end{pmatrix}
= \begin{pmatrix}
8 & v & -1 \\
1 & 6 & 4
\end{pmatrix}.
$$

Найдите значения $(x, y, z)$ и $v$, при которых выражение верно.

Всю матрицу брать не нужно. В соответствии с правилами умножения скалярных значений и матриц, а так же правилами сложения матриц
все значения матрицы в выражении в правой части будут соответствовать значению в правой части после выполнения
всех арифметических действий.  
Например, если рассмотреть значений с индексом $_{1,3}$  
$3*3 +2*(-5) = -1$ -> $9+(-10) = -1$ -> $-1 = -1$

Таким образом можно сразу составить необходимые выражения и найти искомые значения $x,y,z,v$  
$3x+2=8$->$x=2$  
$3y+2(-6)=6$->$y=6$  
$3*4+2z=4$->$z=-4$  
$3*2+2*2=v$->$x=10$
