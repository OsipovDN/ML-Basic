## Задание 1

Написать программу, которая получает на вход строку и возвращает словарь, где:
ключи — символы из этой строки;
значения — количество раз, сколько каждый символ встречается.

## Задание 5
Дана строка текста (или введённая через консоль). Программа должна вернуть словарь с четырьмя ключами:
"гласные",

"согласные",

"цифры",

"пунктуация".

Значения — количество символов каждого типа в строке.

````python
vowels = "аоуыэеёиюя"
result = {"цифры" : 0, "гласные" : 0, "согласные" : 0, "пунктуация" : 0}

in_str = input("Введите текст:",)
for ch in in_str:
    if ch.isspace():
        continue 
    if ch.isdigit():
        result["цифры"] +=1
    elif ch.isalpha():
        if ch.lower() in vowels:
            result["гласные"] +=1
        else:
            result["согласные"] +=1
    else:
        result["пунктуация"] +=1

print(f'"{in_str}" -> "{result}"')
````

## Задание 1: Получение однозначного числа

Пользователь вводит целое число, программа складывает все цифры числа, с полученным числом — то же самое, и так до тех пор, пока не получится однозначное число.

<br>

**Примеры:**
* `545   -> 5`
* `12345 -> 6`

````python
value = input("Введит+ целое число: ",)
in_value = value
if not value.isdigit():
    print(f"Введенное значение {value} не является числом")
while not len(value) == 1:
    sum = 0
    for i in value:
        sum += int(i)
    value = str(sum)
print(f"{in_value} -> {value}")
````

## Задание 2: Кинотеатр

Дан список списков, каждый вложенный список состоит из 1 и 0, количество вложенных списков — количество рядов. Пользователь вводит, сколько билетов ему требуется. Программа должна найти ряд, где можно приобрести нужное количество билетов (места должны быть рядом). Если таких рядов несколько, то ближайший к экрану (ближайшим считается нулевой ряд). Если таких мест нет, то вывести False.

<br>

**Примеры:**
* `[[0,1,1,0], [1,0,0,0], [0,1,0,0]], 2 -> 1`
* `[[0,1,1,0], [1,0,1,0], [1,1,0,1]], 2 -> False`

````python
import random

row = 3     #Колличество рядов
site = 4    #Количество мест в ряду


cinema_hall = [[int(random.choice([True, False])) for _ in range(site)] for _ in range(row)]
try:
    number = int(input("Введите колличество мест: ",))
    result = ''
    if number > len(cinema_hall[0]):
        print("В ряду нет столько мест")
        exit()
    for i in range(row):
        check = False
        for j in range(site):
            if cinema_hall[i][j] == 0:
                if site-j >= number and (not True in cinema_hall[i][j:j + number]):
                    check = True
                    result = i
                    break
        if not result == '':
            break
    print(f"{cinema_hall}, {number} -> {"False" if result == '' else result}")
except ValueError:
    print("Введено некоректное значнеи")
````

## Задание 3: Алгоритм RLE

Необходимо написать упрощенную версию алгоритма RLE. 
Алгоритм RLE объединяет подряд идущие символы в коэффициент и символ.

<br>

**Примеры:**
* `asssdddsssddd -> 1a3s3d3s3d`
* `aaabbbbccccc  -> 3a4b5c`
* `abcba         -> 1a1b1c1b1a`

````python
in_str = input("Введите повторяющиеся подряд символы",)
symbols = "".join(dict.fromkeys(in_str))
result = ''
for ch in symbols:
    result += str(in_str.count(ch)) + ch
print(f"{in_str} -> {result}")
````

## Задание 4: Шифр Цезаря

Реализуйте программу для шифрования текста с помощью шифра Цезаря.  
Шифр Цезаря — это метод шифрования, при котором каждая буква в тексте заменяется буквой, стоящей на фиксированное число позиций дальше в алфавите.

### Требования
1. Программа должна принимать на вход:
   - Строку текста для шифрования
   - Ключ шифра (число от 0 до 25)

2. Правила шифрования:
   - Шифруются только латинские буквы (a-z, A-Z)
   - Сохраняется регистр букв (заглавные остаются заглавными)
   - Пробелы, цифры и знаки препинания остаются без изменений
   - Сдвиг циклический (после 'z' идет 'a', после 'Z' идет 'A')

### Примеры

* `"Hello World", 3      -> "Khoor Zruog"`
* `"Python is great!", 5 -> "Udymts nx lwjfy!"`
* `"XYZ abc", 3          -> "ABC def"`
* `"Test 123", 1         -> "Uftu 123"`

### Алгоритм
* Для каждого символа в строке:
   - Если это латинская буква, найти её позицию в алфавите
   - Добавить к позиции значение ключа
   - Если результат больше 25, использовать остаток от деления на 26
   - Преобразовать обратно в букву, сохранив регистр
   - Если это не буква, оставить символ без изменений


````python
text = input("Введите строку для шифрования:",)
count = int(input("Введите величину сдвига:",))
result = ''
for ch in text:
    if ch.isalpha():
        pos = ord(ch)
        if ch.islower():
            result += chr((pos - ord('a') + count)%26 + ord('a'))
        else:
            result += chr((pos - ord('A') + count)%26 + ord('A'))
    else:
        result +=ch
print(f'"{text}",{count} -> "{result}"')

````

## Задание 5: Табель успеваемости

Пользователь в бесконечном цикле (пока не будет введена пустая строка) вводит строки вида: 'название предмета' 'фамилия ученика' 'оценка'. После окончания ввода программа выводит в консоль название предмета, далее список учеников и все их оценки в виде таблицы.

<br>

**Примеры:**

<br>

*Ввод:*
```python
Математика Иванов  5
Математика Иванов  4
Литература Иванов  3
Математика Петров  5
Литература Сидоров 3
Литература Петров  5
Литература Иванов  4
Математика Сидоров 3
Математика Петров  5
```

<br>

*Результат:*
```python
Математика # вводим название предмета

# выводим список учеников и их оценки
Иванов  5 4
Петров  5 5
Сидоров 3
```

```python
Литература # вводим название предмета

# выводим список учеников и их оценки
Иванов  3 4
Сидоров 3
Петров  5
```

````python
print("Введите предмет/имя/оценку через пробел:")
lessons ={}
while True:
    in_str = input()
    if in_str == '':
        break
    in_str = in_str.split()
    if in_str[0] not in lessons:
        lessons[in_str[0]]={}
    if in_str[1] not in lessons[in_str[0]]:
        lessons[in_str[0]][in_str[1]] = []
    lessons[in_str[0]][in_str[1]].append(in_str[2])

while True:
    lesson = input("Введите название предмета:",)
    if lesson == '':
        break
    for student, grades in lessons[lesson].items():
        print(f"{student}: {', '.join(grades)}")
````
